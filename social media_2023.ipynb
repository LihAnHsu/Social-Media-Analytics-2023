pack all the geo df into all_df_geo list

remove 'Category: Social Networks' and convert first row into column header

```python
all_df_geo = [fb_geo, friendster_geo, google_plus_geo,
              hi5_geo, ig_geo, linkedin_geo,
              myspace_geo, orkut_geo, reddit_geo,
              snapchat_geo, threads_geo, tiktok_geo,
              tumblr_geo, twitter_geo, vine_geo,
              weibo_geo, yt_geo]
```

```python
for i in range(len(all_df_geo)):
    all_df_geo[i].reset_index(inplace=True)
    all_df_geo[i].columns = all_df_geo[i].iloc[0] #replace column header by first row
    all_df_geo[i] = all_df_geo[i].drop(all_df_geo[i].index[0]) #drop first row
    
    # now it look like this: Country | Facebook: (1/1/04 - 7/13/23)
    all_df_geo[i] = all_df_geo[i].astype('str')
    
    #keeping only the part before ':'
    all_df_geo[i].columns = all_df_geo[i].columns.to_series().str.split(':', n = 1).str[0] # n=1, ensures that the split occurs only at the first occurrence of ':' in the string
    
    # fill null with 0, but in here, the nan in second column is object, fillna() can't be apply in here
    all_df_geo[i] = all_df_geo[i].replace('nan', '0')
    
    print(all_df_geo[i])
```
